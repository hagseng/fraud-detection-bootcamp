```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed_data.csv')
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Normal (0) vs Fraud (1) İşlem Sayıları')
plt.yscale('log')
plt.show()
```

```python
print(df['Class'].value_counts(normalize=True))
```

```python
print(df.groupby('Class')['Transaction_Count_Per_Hour'].mean())
```

```python
print(df.groupby('Class')['Hour'].mean())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Class', y='Transaction_Count_Per_Hour', data=df)
plt.title('İşlem Yoğunluğunun Sınıflara Göre Dağılımı')
plt.show()
```
