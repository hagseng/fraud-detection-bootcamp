### EDA - Ham Veri
Feature engineering yapmadan, doğrudan ham veri (Time, Amount, V1-V28) üzerinde analizler gerçekleştirdim.

**Kritik Bulgular:**
1.  **V11**, Class ile en yüksek pozitif korelasyona sahip değişkendir ve diğer V değişkenlerinden bağımsızdır.
2. **V10 ile Class arasındaki korelasyon** çok çok düşüktür. Yani V10 bileşeninin fraud ile ilgili bir bilgi taşımadığını göstermektedir.
3. **V20 ile Amount arasında 0.34'lük belirgin bir pozitif korelasyon** tespit edilmiştir. Bu durum, V20 bileşeninin işlemin parasal büyüklüğüyle ilgili yapısal bir bilgi taşıdığını göstermektedir.

```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Class', data=df)
plt.title('Normal (0) vs Fraud (1) İşlem Sayıları')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==1]['Amount'], bins=30, color='red', label='Fraud', stat="density", common_norm=False)
sns.histplot(df[df['Class']==0]['Amount'], bins=30, color='blue', label='Normal', stat="density", common_norm=False)
plt.title('İşlem Tutarı Dağılımı (Amount)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==1]['Time'], bins=50, color='red', label='Fraud', kde=True, stat="density")
sns.histplot(df[df['Class']==0]['Time'], bins=50, color='blue', label='Normal', kde=True, stat="density")
plt.title('İşlem Zamanı Dağılımı (Saniye Cinsinden)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = df.corr()
sns.heatmap(corr[['Class']].sort_values(by='Class', ascending=False).head(10), annot=True, cmap='coolwarm')
plt.title('Class ile En Yüksek Korelasyona Sahip Özellikler')
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = df.corr()
sns.heatmap(corr[['V11']].sort_values(by='V11', ascending=False).head(10), annot=True, cmap='coolwarm')
plt.title('Class ile En Yüksek Korelasyona Sahip Özellikler')
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = df.corr()
sns.heatmap(corr[['V10']].sort_values(by='V10', ascending=False).head(10), annot=True, cmap='coolwarm')
plt.title('Class ile En Yüksek Korelasyona Sahip Özellikler')
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = df.corr()
sns.heatmap(corr[['V20']].sort_values(by='V20', ascending=False).head(10), annot=True, cmap='coolwarm')
plt.title('Class ile En Yüksek Korelasyona Sahip Özellikler')
plt.show()
```
